````python
big = pd.read_csv('data/songs_18yrs_spotify_data_v5.csv', encoding = "ISO-8859-1")
````

````python
big.columns
````

````python
big['title_lower'] = big['title'].str.lower()
useful_columns = ['year', 'title_lower', 'duration_ms', 'genre', 'broad_genre']
big[useful_columns]
````

````python
hits = pd.read_csv('data/top_10_Rocknroll_by_year.csv', header=None, names=['fulltitle', 'year', 'rank'])
````

````python
#clean up
hits['fulltitle'] = hits['fulltitle'].str.replace('\xa0', '')
hits['title'] = hits['fulltitle'].str.strip().str.split(r'\d\.').str[1].str.split('-').str[0]
hits['artist'] = hits['fulltitle'].str.strip().str.split(r'\d\.').str[1].str.split('-').str[1]
hits['title_lower'] = hits['title'].str.lower()
````

````python
hits['title'][0]
````

````python
songs = pd.merge(hits, big, how='left', on='title_lower')
````

````python
songs.dropna().to_csv('top_songs_cleaned.csv', index=False)
````

````python
songs.shape, hits.shape
````

````python
u2 = ['title_lower', 'year_x', 'rank_x', 'genre', 'broad_genre', 'duration_ms']
df = songs[u2].dropna().groupby('title_lower').first()
````

````python
df2 = df.sort_values(['year_x', 'rank_x'])
````

# Creation of song_dict

```
{2014: {rank:1, duration: 228983, genre:"adsf", broad_genre:'pop'},...,
2015: {rank:1, duration: 228983, genre:"adsf", broad_genre:'pop'},...,
2016: {rank:1, duration: 228983, genre:"adsf", broad_genre:'pop'},...,
2017: {rank:1, duration: 228983, genre:"adsf", broad_genre:'pop'},...,}
```


````python
songs_list = []
for i,row in df2.iterrows():        
    yd = {}
    yd['year'] = row['year_x']
    yd['rank'] = row['rank_x']
    yd['duration'] = row['duration_ms']
    yd['genre'] = row['broad_genre']
    songs_list.append(yd)
````
