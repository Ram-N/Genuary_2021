```python
timings = []
for lat in range(-90,91, 10):
    for lng in range(-180, 181, 10):

        ue = "https://api.sunrise-sunset.org/json?lat="+ str(lat) +  "&lng=" + str(lng) + "&formatted=1"
        response = session.get(ue)
        resp = json.loads(response.html.html)
        try:
            res = resp['results']
            #print(lng, res['sunrise'],res['sunset'] )
            h,m,s = res['sunrise'].split(':')
            sr = int(h)*60  + int(m) + (round(int(s.split(' ')[0])/60))
            if int(h) == 12 and s.split(' ')[1] == 'AM':
                sr -=720
            elif s.split(' ')[1] == 'PM' and int(h) != 12:
                sr +=720
                
            h,m,s = res['sunset'].split(':')
            ss = int(h)*60  + int(m) + (round(int(s.split(' ')[0])/60))
            if int(h) == 12 and s.split(' ')[1] == 'AM':
                ss -=720
            elif s.split(' ')[1] == 'PM' and int(h) != 12:
                ss +=720

            dur = ss-sr
            if dur < 0:
                dur+=1440
            timings.append(((lat, lng), sr, ss, dur))

        except:
            pass
```

```python
td1 = {}

for tset in timings:
    ll, sr, ss, dur = tset
    if ll in td1:
        td1[ll].append((sr, ss, dur))
    else:
        td1[ll] = [(sr, ss, dur)]
```
